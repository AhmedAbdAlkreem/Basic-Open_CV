```python
# To convert image from BGR to RGB.
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# To change the area of image but size of image did not change.
plt.figure(figsize = (5, 5))

# To show rgb image.
plt.imshow(rgb_img)
plt.show()
```

# Convert Image To GRAY

# Helper Function

```python
# To show image. (GRAY OR RGB)
def imshow(img, figsize = (10, 10)):
    plt.figure(figsize = figsize)
    plt.imshow(img, cmap ='gray')
    plt.show()
```

```python
# Read the image.
img = cv2.imread('sudoku.jpg')

# Convert image to GRAY.
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

imshow(gray_img)
```

```python
# Blurring image reduce noise.
gray_img = cv2.blur(gray_img, ksize = (5, 5))

imshow(gray_img)
```

```python
# Show the type of image.
gray_img.dtype
```

# Sobel Detection

## Sobel (Vertical Edges):

```python
# Using sobel to detect the edges.
# dx = 1 ==> verticle edges.
# dy = 1 ==> horizontel edges.
img_sobel_x = cv2.Sobel(gray_img, cv2.CV_8U, dx = 1, dy = 0, ksize = 5)

# show the image before sobel.
imshow(img_sobel_x)
```

## Sobel (Horizontal Edges):

```python
# Using sobel to detect the edges.
img_sobel_y = cv2.Sobel(gray_img, cv2.CV_8U, dx = 0, dy = 1, ksize = 5) # dy = 1 ==> horizantel edges

imshow(img_sobel_y)
```
